```python
mu = 0
sigma = 1
x = np.sort(df['REGION_POPULATION_RELATIVE'])

y_pdf = ss.norm.pdf(x, mu, sigma) # the normal pdf
y_cdf = ss.norm.cdf(x,  mu, sigma) #

plt.plot(x, y_pdf, label='pdf')
plt.plot(x, y_cdf, label='cdf')
plt.legend();
```

```python
# 繪製 Empirical Cumulative Density Plot (ECDF)

counts, bin_edges = np.histogram (df.AMT_INCOME_TOTAL, bins=100, normed=True)
cdf = np.cumsum (counts)
plt.plot (bin_edges[1:], cdf/cdf[-1])
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python



x, y = ecdf(df["REGION_POPULATION_RELATIVE"])

plt.figure(figsize=(8,7))

sns.set()


samples = np.random.normal(np.mean(df["REGION_POPULATION_RELATIVE"]),
                           np.std(df["REGION_POPULATION_RELATIVE"]), 
                           size=100000)

x_theor, y_theor = ecdf(samples)

plt.plot(x_theor, y_theor)
plt.plot(x, y, marker=".", linestyle="none")

plt.legend(('Normal Distribution', 'Empirical Data'), loc='lower right')
```
