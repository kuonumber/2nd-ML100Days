- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
```

```python
kmeans  = KMeans(n_clusters=5, random_state=1)
```

```python
kmeans.fit_transform(X)
```

```python
estimators = [('k3', KMeans(n_clusters=3, random_state=1)),
               ('k5', KMeans(n_clusters=5, random_state=1)),
                ('k8', KMeans(n_clusters=8, random_state=1)),
               ('k8_bad_init', KMeans(n_clusters=8, n_init=1, init='random'))
                 ]
```

```python
fignum = 1
titles = ['3 clusters', '5 clusters', '8 clusters', 'k8_bad_init']
for name, est in estimators:

    fig = plt.figure(fignum, figsize=(10, 7))
    ax = Axes3D(fig, rect=[0, 0, .8, 1], elev=48, azim=134)

    ## fit data
    est.fit(X)
    labels = est.labels_

    ax.scatter(X[:, 3], X[:, 0], X[:, 2],
               c=labels.astype(np.float), edgecolor='k')

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1
```
